```python
# [Importing the Required Libraries]
import pandas as pd
import numpy as nu
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
# [Generate randon numbers]
randomNumber = nu.random.rand(20)
```

```python
# [Printing the random numbers]
print(randomNumber)
```

```python
# [Selecting the Style of the Plot]
style.use('ggplot')
```

```python
# [Plotting the Random Numbers]
plt.plot(randomNumber,'g',label='line one',linewidth=2)
# [X-Axis Stands for the Number of Random Numbers]
plt.xlabel('Range')
# [Y-Axis Stands for the Actual Random Number]
plt.ylabel('Numbers')
# [Title of the Plot]
plt.title('Random Number Plot')
```

```python
# [Data from Ohio State, leading cause of death in the state from the year 2012]
# cause of death
cause = 'Chronic Diseases', 'Unintentional Injuries', 'Alzheimers', 'Influenza and Pnemonia', 'Sepsis', 'Others'
# [Percentile]
percentile = [40,15,15,5,10,25]
```

```python
# [Set the pie chart plot properties]
# [Set figure size]
plt.figure(figsize = (10,10))
```

```python
# [Explode the largest pie in the dataset]
explode = (0.05,0,0,0,0,0)
```

```python
# [Set pie chart properties]
plt.pie(percentile, labels=cause,explode=explode,autopct='%1.1f%%',startangle=70)
# [Set axis equal to draw pie as circle]
plt.axis('equal')
# [Adding a title to the Plot]
plt.title('Ohio State - 2012 : Leading Causes of Death')
# [Showing the Plot]
plt.show()
```
